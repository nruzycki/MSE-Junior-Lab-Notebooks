```python
#bringing in the Python Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#bringing in the .csv files
#here we have three dogbone files and one file with the virgin polymer for this example
dfv=pd.read_csv('QV.csv')
df1=pd.read_csv('Qth1.csv')
df2=pd.read_csv('Qth2.csv')
df3=pd.read_csv('Qth3.csv')

#Name the columns for each dataframe
dfv.columns =['Wavenumber_V', 'CtsV'] 
df1.columns =['Wavenumber_1', 'Cts1']
df2.columns =['Wavenumber_2', 'Cts2']
df3.columns =['Wavenumber_3', 'Cts3']
```

```python
plt.scatter(x=dfv['Wavenumber_V'], y=dfv['CtsV'])
plt.xlabel("Wavenumber(cm-1)")
plt.ylabel("Arb. Units")
plt.title("Virgin Material FTIR")
plt.show()
```

```python
#How to plot the data and show the x-axis reversed (as traditonal in FTIR) 

x=dfv['Wavenumber_V']
y=dfv['CtsV']

fig, (ax1, ax2) = plt.subplots(1, 2) #this tells the plot that you want two plots in line

ax1.plot(x, y)
ax1.set_title("Normal Plot")
ax1.set_xlabel('Wavenumber(cm-1)')
ax1.set_ylabel('Arb. Units')

ax2.plot(x, y)
ax2.set_title("Reverted axes")
ax2.invert_xaxis()
ax2.set_xlabel('Wavenumber(cm-1)')
ax2.set_ylabel('Arb. Units')

fig.tight_layout()
plt.show()

#The last program you ran for tensile showed you how to save a figure to put in a paper. Find this function and paste here to save your figure file.
```
